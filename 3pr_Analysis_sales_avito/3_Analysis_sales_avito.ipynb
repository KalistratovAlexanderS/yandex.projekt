## Предобработка данных

Ранее я обнаружил, что  12180 объявлений содержат пропусти в выcоте потолка, а также почти треть объявлений имеют данные с завышенными, по сравнению с высотой по СНиП, значениям потолков также есть ряд помещений где высота сильно завышена более 20 м

```python
#заменим все пропуски на медиану, 
height_median = df['ceiling_height'].median()
df['ceiling_height'] = df['ceiling_height'].fillna(height_median)
df['ceiling_height'].isna().value_counts()
```

Число пропущенных этажностей дома не возможно определить исходя из данных, но чтобы не терять данные заменим на медиану по населенному пункту, чтобы не было дробных зчачений округлим данные

```python
#удаляем строки, где количество комнат указано как 0, так как их количество менее 1%
df.drop(df[df['rooms'] == 0].index, inplace=True)

# также поступим с квартирами где не указана этажность
df.drop(df[df['floors_total'].isnull()].index, inplace=True)
```

```python
#заменим на медиану высоты в населенном пункт
df = df.dropna(subset=['locality_name']).reset_index(drop=True)
df['floors_total'] = df['floors_total'].astype('int')
floors_total_median = dict(df.pivot_table(index = 'locality_name', values = 'floors_total', aggfunc = 'median')['floors_total'])
type(floors_total_median)
def floors_total_fun(row):
    loc_name = row['locality_name']
    fl_total = row['floors_total']
    if fl_total == 0:
        for key, value in floors_total_median.items():
            if key in loc_name:
                return value
    
    return fl_total
df['floors_total'] = df.apply(floors_total_fun, axis=1).round()
```

### Цена квадратного метра, дата публикации, метка-этажность

```python
#в начале приведем столбец first_day_exposition к типу данных datetime
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y%m%dT%H:%M:%S')
df['post_weekday'] = df['first_day_exposition'].dt.weekday
df['post_month'] = df['first_day_exposition'].dt.month
df['post_year'] = df['first_day_exposition'].dt.year   
```

```python
#Напишем функцию для определения этажа и применим её к столбцу floor

floors_total = df['floors_total']

def find_floor(row):
    floor = row['floor']
    if floor == 1:
        return('первый')
    elif floor == row['floors_total']:
        return('последний')
    return('другой')        
        
df['floor_type'] = df.apply(find_floor, axis=1) 
```
